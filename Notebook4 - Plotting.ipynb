## <a id='toc1_1_'></a>[Matplotlib](#toc0_)

Matplotlib ist die standard Python-Bibliothek zur Erstellung von 
publikationsreifen Plots. Sie bietet eine Vielzahl von Funktionen zum 
Erstellen von verschiedenen Arten von Diagrammen wie Linien-, Balken- 
und Flächendiagrammen sowie Histogramme, Streudiagramme und mehr. Mit 
Matplotlib können Benutzer auch benutzerdefinierte Diagramme und 
Anpassungen vornehmen, um ihre Darstellungen zu optimieren. Die 
Bibliothek ist sehr flexibel und kann nahtlos mit anderen 
Python-Bibliotheken wie NumPy und Pandas integriert werden.

```python
from matplotlib import pyplot as plt
import numpy as np
```

Daten erstellen:

```python
alter = [range(25, 35)] # Note: range(25, 35) wird nicht unpacked
alter = [*range(25, 35)] # range(25, 35) wird unpacked
```

Einschub: 
\* ist ein unpacking operator in python

Der \* Operator unpacked die Elemente von iterable objects (=Objekte 
über deren Elemente man mit einem for-loop iterieren kann z.B. lists, 
tuples, strings..) 

```python
salary = np.around(np.random.normal(50000, 10000, 10), 0).tolist()
```

Erster einfacher Plot:

```python
plt.plot(alter, salary)
plt.show
```

Beschriftung hinzufügen:

```python
plt.plot(alter, salary)
plt.xlabel("Alter")
plt.ylabel("Medianeinkommen")
plt.title("Medianeinkommen über Alter")
plt.show
```

Verschiedene Gruppen plotten:

```python
# Erstellung von weiteren Einkommensdaten:
salary_dt = salary
salary_sw = np.around(np.random.normal(60000, 5000, 10), 0).tolist()

# Plot beider Einkommensgruppen

plt.plot(alter, salary_dt, label="Deutschland")

plt.plot(alter, salary_sw, label="Schweiz")

plt.xlabel("Alter")
plt.ylabel("Medianeinkommen")
plt.title("Medianeinkommen über Alter")

plt.legend() # fügt die Legende auf Basis der Labels hinzu

plt.show
```

Modifizierung der Darstellung:

```python
plt.plot(alter, salary_dt, label="Deutschland",
        color="pink", marker =".")

plt.plot(alter, salary_sw, label="Schweiz",
        linestyle = "--")

plt.xlabel("Alter")
plt.ylabel("Medianeinkommen")
plt.title("Medianeinkommen über Alter")

plt.legend() # fügt die Legende auf Basis der Labels hinzu
plt.grid(True) # fügt Gridlines hinzu
plt.show
```

### <a id='toc1_2_5_'></a>[Barplot (mit Anpassungen)](#toc0_)

Oft müssen Plots Datenstrukturen darstellen die über 
Standardarstellungen hinaus gehen. Wie genau eine bestimmte Darstellung 
erreicht werden kann hängt natürlich immer vom Einzellfall ab. Als 
Beispiel schauen wir uns einen Barplot an. Wir verwenden dazu den schon  
bekannten Datensatz zu Leihfahrädern 
([Link](https://www.kaggle.com/competitions/bike-sharing-demand/data)).

```python
import pandas as pd
bike_sharing = pd.read_csv("data/day.csv", sep = ",")
```

```python
# Data cleaning (wie in der letzten Session):
bike_categories = ['season', 'holiday', 'weekday',
                    'workingday', 'weathersit']

for variable in bike_categories:
    bike_sharing[variable] = bike_sharing[variable].astype("category")

bike_sharing
```

```python
bike_sharing['season'] = bike_sharing['season'].map({1: 'winter',
                                                    2: 'spring',
                                                    3: 'summer',
                                                    4: 'autumn'})
bike_sharing
```

Wir wollen für die vier Jahreszeiten die Fahrrad Ausleihzahlen von 
casual und registrierten Nutzern gegenüberstellen.

```python
# X-Achse:
season = bike_sharing["season"].cat.categories
print(season)
```

```python
# Y-Achse:

casual = bike_sharing.groupby("season").sum()["casual"]
registered = bike_sharing.groupby("season").sum()["registered"]
casual
```

```python
counts = pd.concat([casual, registered], axis=1)
counts
```

Wenn wir die Plots übereinander legen erreichen wir nicht den 
geschünschten Effekt einer Gegenüberstellung. Die Bars werden 
aufeinander gelegt:

```python
plt.bar(x = season, height = counts["registered"])
plt.bar(x = season, height = counts["casual"])
plt.xlabel('Jahreszeit')
plt.ylabel('Ausleihzahlen')
plt.title('Histogram')

# Print the chart
plt.show()
```

Um die Bars voneinander zu trennen müssen wir etwas in die Trickkiste 
greifen:

```python
x_indexes = np.arange(len(season))
width = 0.25
loc = 0.25/2

with plt.xkcd():

        plt.bar(x = x_indexes - loc,
                height = counts["registered"],
                color = "purple",
        width = width)
        plt.bar(x = x_indexes + loc,
                height = counts["casual"],
                color = "green", 
                width=width)
        
plt.xlabel('Jahreszeit')
plt.ylabel('Ausleihzahlen')
plt.title('Histogram')

plt.xticks(ticks=x_indexes, labels=season)

```

Das Seaborn package ist eine Erweitung bzw. Alternative Matplotlib die 
auf Matplotlib basiert. Seaborne funktioniert besser mit pandas data 
frames. Die Seaborn syntax ist im gleich zu matplotlib oft kürzer und 
einfacher. 

Seaborn Beispiel: Histogram + Kernel Density Estimation (KDE):

```python
sns.displot(data=dataset, x='petal length (cm)', bins=10, kde=True)
```

Will man eine Normalverteilung mit Parametern basierend auf den Daten 
hinzufügen muss man etwas kreativ werden:

```python
from scipy import stats

def normal(mean, std, color="black"):
    x = np.linspace(mean-3*std, mean+3*std, 200) # erstelle X-Koords
    p = stats.norm.pdf(x, mean, std) # Norm Density Werte der X-Koords
    z = plt.plot(x, p, color, linewidth=2) # Verbinde Density Werte
```

```python
mean = dataset["petal length (cm)"].mean()
std = dataset["petal length (cm)"].std()

dataset["sd"] = (dataset[
    "petal length (cm)"] - mean) / std

sns.displot(data=dataset,
            x="sd",
            kde=True, stat = "probability", discrete = True)
normal(0, 1)
```

### <a id='toc1_4_3_'></a>[Advanced Plotting: 3D Vectors in plotly](#toc0_)

Das Plotten komplizierer Daten erfodert eine tiefere Auseinandersetzung 
mit der Syntax von plotly. Im folgenden Beispiel werden wir die 
Basisvektoren einer 3x3 Matrix zu plotten.

Angenommen wir haben folgende Matrix:

$\begin{pmatrix}
 2 & 1 & 0 \\
 1 & 1 & 1 \\
 3 & 2 & 2 \\
\end{pmatrix}$

Wie können wir plotly nutzen um die 3 Basisvektoren:

$a = \begin{pmatrix}2 \\ 1 \\ 3 \\\end{pmatrix}$; 
$b = \begin{pmatrix}1 \\ 1 \\ 2 \\\end{pmatrix}$; 
$c = \begin{pmatrix}0 \\ 1 \\ 2 \\\end{pmatrix}$

plotten?

```python
import numpy as np
```

```python
matrix = np.array([[2,1,0],[1,1,1],[3,2,2]])
```

```python
a = matrix[:,0]
b = matrix[:,1]
c = matrix[:,2]
```

Ursprung (hier Nullvektor) erstellen:

```python
Nv = np.array([0,0,0])
```

Koordinaten auf Basis der Vektoren bestimmen

```python
Coords = np.stack([Nv,a,Nv,b,Nv,c], axis = 1)
```

```python
x = Coords[0].tolist() # Liste der X-Koords aller Punkte
y = Coords[1].tolist() # Liste der Y-Koords aller Punkte
z = Coords[2].tolist() # Liste der Z-Koords aller Punkte
```

```python
pairs = [(0,1), (2,3),(4,5)]
pairs
```

Folgender Code zieht die Startkoordinaten der Vektoren aus den 
Koordinatenvektoren heraus:

```python
x_m = [x[p[0]] for p in pairs]
y_m = [y[p[0]] for p in pairs]
z_m = [z[p[0]] for p in pairs]
```

```python
trace1 = go.Scatter3d(
    x=x_m, #X-Koords der Scatterpunkte
    y=y_m, #Y-Koords der Scatterpunkte
    z=z_m, #Z-Koords der Scatterpunkte
    mode='markers',
    name='markers',
    line=dict(color='green'),
    marker=dict(size = 2)
)

# fig = go.Figure(trace1) # trace1 (3 überlapp. Punkte im Coord Ursp.)
# fig.show() # trace1 anschauen
```

```python
# Initiere Listen für die Vektorlinien:
x_lines = list()
y_lines = list()
z_lines = list()

# Speichere alle x,y,z coords in einer liste, füge nach jedem 
# Punktepärchen ein "None" hinzu:

for p in pairs: # (0,1), (2,3), (4,5) # Wählt Punktepärchen aus
    for i in range(2): # 0, 1 # Wählt ersten, dann zweiten Punkt
        x_lines.append(x[p[i]])
        y_lines.append(y[p[i]])
        z_lines.append(z[p[i]])

print(x_lines)
```

Mit den "lines" Listen initialisieren wir jetzt eine trace für die 
Vektorlinien:

```python
trace2 = go.Scatter3d(
    x=x_lines,
    y=y_lines,
    z=z_lines,
    mode='lines',
    line = dict(width = 2, color = 'rgb(255, 0,0)')
)

# fig = go.Figure(trace2) 
# fig.show() # trace2
```

Kurzer Blick was wir bisher geschafft haben:

```python
fig = go.Figure(data=[trace1, trace2])
fig.show()
```

Die Spitzen der Vektoren fehlen noch. Die 
<code>go.Cone()</code> trace könnte hier passend sein. Dazu müssen wir 
zunächst verstehen wie die Cone trace funktioniert. In der 
Dokumentation finden wir folgendes Statement:

"Use cone traces to visualize vector fields. Specify a vector field 
using 6 1D arrays, 3 position arrays x, y and z and 3 vector component 
arrays u, v, w. The cones are drawn exactly at the positions given 
by x, y and z."

Wir starten mit einer einfachen Cone um herauszufinden ob sich alles 
so verhält wie wir uns das vorstellen:

```python
cone_test = go.Figure(data=go.Cone(x=[0], y=[0], z=[0],
                                    u=[1], v=[0], w=[0]))

cone_test.show()
```

Wir stellen ein Problem fest: es scheint als würde die Cone nicht auf
den Koordinaten (0,0,0) beginnen (Boden) oder aufhören (Spitze).
Wir versuchen die Problematik zu verdeutlichen:

```python
cone_test = go.Figure(data=go.Cone(x=[0], y=[0], z=[0],
                                    u=[1], v=[0], w=[0],
                                    opacity = 0.5))

cone_test.add_trace(go.Scatter3d(
    x = [0], y = [0], z = [0],
    marker=dict(color = "black")
))

cone_test.show()
```

Ein weiterer Blick in die Dokumentation offenbart, dass der Parameter
"anchor" den Punkt der Cone referenziert der durch x,y,z fixiert wird.
Default ist hier "cm" (center of mass), wir stellen auf "tail" um:

```python
cone_test = go.Figure(data=go.Cone(x=[0], y=[0], z=[0],
                                    u=[1], v=[0], w=[0],
                                    opacity = 0.5, 
                                    anchor="tail"))

cone_test.add_trace(go.Scatter3d(
    x = [0], y = [0], z = [0],
    marker=dict(color = "black")
))

cone_test.show()
```

Das sieht besser aus. Wir erkennen jetzt jedoch, dass die Länge der 
Cone nicht 1 ist, obwohl der Vektorfeld Vektor (1,0,0) ist. Ein 
weiterer Blick in die Dkoumentation zeigt, dass die Länge der Cone 
standardisiert ist. Wir ändern die entsprechenden Parameter:

```python
cone_test = go.Figure(data=go.Cone(x=[0], y=[0], z=[0],
                                    u=[1], v=[0], w=[0],
                                    opacity = 0.5, 
                                    anchor="tail",
                                    sizemode = "absolute",
                                    sizeref=1))

cone_test.add_trace(go.Scatter3d(
    x = [0], y = [0], z = [0],
    marker=dict(color = "black")
))

cone_test.add_trace(go.Scatter3d(
    x = [0,1], y = [0,0], z = [0,0],
    mode = "lines",
    line = dict(width = 3, color = 'black')
))

cone_test.show()
```

Die Cone verhält sich nun genauso wie erwartet. Mit dem nun gewonnenen
Verstädnis über go.Cone() funktion, können wir unsere Vektoren 
vervollständigen: 

```python
arrow_tip_ratio = 0.1
arrow_starting_ratio = 0.98

x = Coords[0].tolist() # Liste der X-Koords aller Punkte
y = Coords[1].tolist() # Liste der Y-Koords aller Punkte
z = Coords[2].tolist() # Liste der Z-Koords aller Punkte

# Der loop nimmt die Startkoordinaten jeder Dimension eines jeden 
# Vektors (e.g.: x[p[0]]) und addiert die 98% der Länge. Somit starten
# die Cones 2% vor der Spitze des Vektors.
    
# Der zweite teil des loop skaliert die Vektoren runter auf 0.1%, für 
# das Vektorfeld der Cones.

for p in pairs:
    fig.add_trace(go.Cone(
        x=[x[p[0]] + arrow_starting_ratio*(x[p[1]] - x[p[0]])],
        y=[y[p[0]] + arrow_starting_ratio*(y[p[1]] - y[p[0]])],
        z=[z[p[0]] + arrow_starting_ratio*(z[p[1]] - z[p[0]])],
        u=[arrow_tip_ratio*(x[p[1]] - x[p[0]])],
        v=[arrow_tip_ratio*(y[p[1]] - y[p[0]])],
        w=[arrow_tip_ratio*(z[p[1]] - z[p[0]])],
        showlegend=False,
        showscale=False,
        autocolorscale=False,
        colorscale=[[0, 'rgb(255,0,0)'], [1, 'rgb(255,0,0)']],
        anchor="tail",
        sizemode = "absolute",
        sizeref=0.1
        ))

fig.show()
```

Das sieht gut aus. Abschließend wollen wir alles in einer Funktion 
zusammenfügen und die Farben ändern:

```python
def Basisvectors(m,
                arrow_tip_ratio,
                arrow_starting_ratio):

    fig = go.Figure()

    a = m[:,0]
    b = m[:,1]
    c = m[:,2]

    Nv = np.array([0,0,0])

    Coords = np.stack([Nv,a,Nv,b,Nv,c], axis = 1)

    x = Coords[0].tolist() # Liste der X-Koords aller Punkte
    y = Coords[1].tolist() # Liste der Y-Koords aller Punkte
    z = Coords[2].tolist() # Liste der Z-Koords aller Punkte

    pairs = [(0,1), (2,3),(4,5)]

    x_m = [x[p[0]] for p in pairs]
    y_m = [y[p[0]] for p in pairs]
    z_m = [z[p[0]] for p in pairs]

    trace1 = go.Scatter3d(
        x=x_m, #X-Koords der Scatterpunkte
        y=y_m, #Y-Koords der Scatterpunkte
        z=z_m, #Z-Koords der Scatterpunkte
        mode='markers',
        name='markers',
        line=dict(color='black'),
        marker=dict(size = 2)
    )

    fig.add_trace(trace1)

    col = ["green", "none", "red", "none", "blue", "none"]

    x_lines = list()
    y_lines = list()
    z_lines = list()

    for p in pairs: # (0,1), (2,3), (4,5) # Wählt Punktepärchen aus
        for i in range(2): # 0, 1 # Wählt ersten, dann zweiten Punkt
            x_lines.append(x[p[i]])
            y_lines.append(y[p[i]])
            z_lines.append(z[p[i]])

    for p in pairs:
        trace2 = go.Scatter3d(
            x=x_lines[p[0]:(p[1]+1)],
            y=y_lines[p[0]:(p[1]+1)],
            z=z_lines[p[0]:(p[1]+1)],
            opacity=0.5,
            mode='lines',
            line = dict(width = 2, color = col[p[0]])
        )
        fig.add_trace(trace2)
    
    for p in pairs:
        fig.add_trace(go.Cone(
            x=[x[p[0]] + arrow_starting_ratio*(x[p[1]] - x[p[0]])],
            y=[y[p[0]] + arrow_starting_ratio*(y[p[1]] - y[p[0]])],
            z=[z[p[0]] + arrow_starting_ratio*(z[p[1]] - z[p[0]])],
            u=[arrow_tip_ratio*(x[p[1]] - x[p[0]])],
            v=[arrow_tip_ratio*(y[p[1]] - y[p[0]])],
            w=[arrow_tip_ratio*(z[p[1]] - z[p[0]])],
            showlegend=False,
            showscale=False,
            autocolorscale=False,
            colorscale=[[0, col[p[0]]], [1, col[p[0]]]],
            anchor="tail",
            sizemode = "absolute",
            sizeref=0.1,
            opacity=0.5
            ))
    
    fig.show()

```
